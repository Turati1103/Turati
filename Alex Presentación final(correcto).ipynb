```python
import seaborn as sns
import numpy as np 
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd
```

```python
df = pd.read_csv('cbb.csv')
df
```

## Pregunta 1: Existe alguna correlacion entre W(Partidos ganados) y EFG_O(Eficencia de tiro al momento de atacar)

```python
sns.lmplot(x = 'EFG_O',
           y = 'W',
           data = df,
           fit_reg = False)
plt.title(' EFG_O VS W', fontsize = 15)
```

Aqui lo que podemos ver es que si esxiste una correlacion lineal positiva entre los juegos ganados y la eficencia de tiro al momento de atacar, esto quiere decir que si mejoramos la efeicencia de tiro al momento de atacar podremos ganar mas partidos y por lo tanto ser rankeado como un mejor equipo

## Pregunta 2:Cual es el rango de 3P_O (Porcentaje de triples a distancia al momento de atacar)?


```python
sns.boxplot(x = df['TEAM'], y= df['3P_O'],
             whis = (0,100)). set_title("Porcentaje de tripels")
plt.xticks(rotation=90)
```

Como podemos observar tenemos muchisimos datos ya que tenemos muchos equipos pero podemos ver que el porcentaje de triples por equuipo va cambiando, al ver la grafica podemos ver que el rango de triples anotados al momento de atacar puede llegar desde los 25 a los 45, esto en el basketball es muy importante ya que la eficacia de tus tiros determinara cuantos puntos anote sy esto puede hacer que ganes o pierdas partidos. 

## Pregunta 3 : Cual es el porcentaje de (3P_D) Porcentaje de triples a distancia al momento de defender?

```python
sns.boxplot(x = df['TEAM'], y= df['3P_D'],
             whis = (0,100)). set_title("Porcentaje de tripels")
plt.xticks(rotation=90)
```

Aqui podemos ver los porcentajes de puntos que cada equipo permite al defender, aqui tenemos que hacer una comparación con la tabla de arriba ya que si permites que te metan mas de los que tu metes estariamos en serios problemas si queremos ganar algun juego, esta tabla nos ayudara mucho a ver que equipo defeienden bien e area de tres ya que lo que se sta tomadno en cuenta es el trio de tirple.

```python
sns.lmplot(x = '3P_D',
           y = '3P_O',
           data = df,
           fit_reg = False)
plt.title(' EFG_O VS W', fontsize = 15)
```

Como podemos ver el 3P_D y el 3P_O no tienen correlacion ya que uno no afecta al otro ya que uno es cuantos puntos de 3 puedes meter y el otro es cuantos puntos de tres te meten claro estos dos si tendran correlación conlas victorias y las derrotas pero entre ellas no existe la correalción 

## Pregunta 4: Cuales fueron los 5 equipos que mas ganaron partidos ? 

```python
pivot_MM = pd.pivot_table(df,index= ['TEAM'],values = ['W'])
pivot_MM = pivot_MM.sort_values('W', axis=0,ascending=False)
pivot_MM[:5]
```

```python
first_5 = pivot_MM[:5]
first_5.plot(kind='bar')
```

Aqui lo que queriamos hacer fue sacar las universidades que han agando mas partidos en todos los años ya que tenemos una base de datos muy grande a veces no se puede apreciar el comportmaieno aprte queriamos ver cuales eran las mejores 5 universidades solo tomando en cuenta los partidos ganados y nos dimos cuenta que Gonzaga es la mejro escuela solo tomando en cuenta los partidos ganados

## Pregunta 5: Rango de BARTHAG entre los primeros 5 equipos

```python
pivot_MM = pd.pivot_table(df,index= ['TEAM'],values = ['BARTHAG'])
pivot_MM = pivot_MM.sort_values('BARTHAG', axis=0,ascending=False)
p5 = (pivot_MM[:5])
```

```python
tabla = sns.lineplot(data= p5)
tabla.set_xlabel('Equipos',fontsize = 15)
tabla.set_ylabel('Barthag',fontsize = 15)
```

El barthag es un pronostico de victoria contra un equipo de primer adivisión asi que en este caso solo marcamos los mejores 5 equipos ya que si poniamosa  todos no iba a entenderse la grafica, podemos ver tiene una pendiente negativa y esto tiene mucho sentido ya que si estos son los equipos que mas ganaron eentonces tendran mas oportunidad de ganar contra equipos de primera division, tambien podemos decir que mientra mas avanca va callendo mas es decir que entre mas lejos estes de los primeros 5 sera mas dificil poder ganarle a un equipo de primera division
